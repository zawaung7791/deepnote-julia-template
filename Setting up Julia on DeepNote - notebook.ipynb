## <center>Julia kerenel and how to setup + Intro to Julia</center>

```
FROM deepnote/python:3.7

RUN wget https://julialang-s3.julialang.org/bin/linux/x64/1.4/julia-1.4.2-linux-x86_64.tar.gz && \
    tar -xvzf julia-1.4.2-linux-x86_64.tar.gz && \
    mv julia-1.4.2 /usr/lib/ && \
    ln -s /usr/lib/julia-1.4.2/bin/julia /usr/bin/julia && \
    rm julia-1.4.2-linux-x86_64.tar.gz && \
    julia  -e "using Pkg;pkg\"add IJulia\""
    
ENV DEFAULT_KERNEL_NAME "julia-1.4"
```
